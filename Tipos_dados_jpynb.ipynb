```python
#tipos de dados no python, revisão

#int

idade = 17
print("idade")
print(idade)
print(type(idade))

#valores (ponto flutuante) valores com virgula (ponto separador de decimais)
altura = 1.85
print("altura")
print(altura)
print(type(altura))


#str (string) texto
nome = "Askii"

#bool - booleano (sim,nao,verdade)
#(mentira,ligado,desligado) ou seja, só possui duas opções de escolhas a se fazer
briga = False
print("briga")
print(briga)
print(type(briga))


#list Lista (coleção ordenada)
frutas = ["blueberry","morango","manga"]
print("frutas")
print(frutas)
print(type(frutas))

#tuple - dupla, como por exemplo, coordenadas
coodernadas = (52.640, 24.325)

#set - conjunto
numeros = {1,2,2,3,3,4,5,5}
print("numeros")
print(numeros)
print(type(numeros))

#dict - dicionario
pessoas = {"nome": "Askii" , "idade": 17}
print("pessoas")
print(pessoas)
print(type(pessoas))

#None Typee = quando não tem nada, mas preciso informar um valor
valor = None
print("valor")
print(valor)
print(type(valor))
```
